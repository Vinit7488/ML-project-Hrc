## q17.Create a new column in the existing dataframe as “'amount_in_usd'” and convert all the non-USD currencies in USD and store them in the same column. 

```python
df["ORDER_AMOUNT"]=pd.to_numeric(df["ORDER_AMOUNT"])
```

```python
def converter(curr,num):
    if curr=="USD":
        return num
    elif curr=="EUR":
        return num*1.08
    elif curr=="AUD":
        return num*0.65
    elif curr=="CAD":
        return num*0.74
    elif curr=="GBP":
        return num*1.24
    elif curr=="MYR":
        return num*0.22
    elif curr=="PLN":
        return num*0.24
    elif curr=="AED":
        return num*0.27
    elif curr=="HKD":
        return num*0.13
    elif curr=="CHF":
        return num*1.10
    elif curr=="RON":
        return num*0.22
    elif curr=="SGD":
        return num*0.74
    elif curr=="CZK":
        return num*0.045
    elif curr=="HU1":
        return num*0.0028916
    elif curr=="NZD":
        return num*0.61
    elif curr=="BHD":
        return num*2.65
    elif curr=="SAR":
        return num*0.27
    elif curr=="QAR":
        return num*0.27
    elif curr=="KWD":
        return num*3.25
    else:
        return num*0.093
        
```

```python
##df["amount_in_usd"]= df[['ORDER_CURRENCY','ORDER_AMOUNT']].apply(lambda df: converter(df["ORDER_CURRENCY"],df["ORDER_AMOUNT"]),axis=1)
df["amount_in_usd"]=np.vectorize(converter)(df["ORDER_CURRENCY"],df["ORDER_AMOUNT"])
```

```python
# Finding the IQR
percentile25 = df['amount_in_usd'].quantile(0.25)
percentile75 = df['amount_in_usd'].quantile(0.75)
```

```python
iqr = percentile75 - percentile25
```

```python
upper_limit = percentile75 + 1.5 * iqr
lower_limit = percentile25 - 1.5 * iqr
```

```python
print("Upper limit",upper_limit)
print("Lower limit",lower_limit)
```

```python
outliers = (df['amount_in_usd']<lower_limit) | (df['amount_in_usd']>upper_limit)
print(outliers.sum())
```

## As we can see the distribution is highly right skewed firstly we need to perform some log transformation to deal with the outliers to get better results

```python
from sklearn.preprocessing import PowerTransformer
from sklearn.preprocessing import FunctionTransformer
```

```python
log_transform = FunctionTransformer(np.log1p, validate=True)

# Apply the log transformation to the "amount_in_usd" column
df["amount_in_usd_log"] = log_transform.transform(df[["amount_in_usd"]])

# Print the transformed DataFrame
print(df)
```

```python
plt.figure(figsize=(16,5))
plt.subplot(121)
sns.kdeplot(df['amount_in_usd'])

plt.figure(figsize=(16,5))
plt.subplot(122)
sns.kdeplot(df['amount_in_usd_log'])
```

```python
# Finding the IQR
q1 = df['amount_in_usd_log'].quantile(0.25)
q3 = df['amount_in_usd_log'].quantile(0.75)
iqr=q3-q1
```

```python
upper_limit = q3 + 1.5 * iqr
lower_limit = q1 - 1.5 * iqr
```

```python
print("Upper limit",upper_limit)
print("Lower limit",lower_limit)
```

```python
fig=plt.figure(figsize=(15,10),dpi=200)
box=sns.boxplot(df["amount_in_usd_log"])
```

```python
len(df[df['amount_in_usd_log']  > 12.5])
```

## here we will perform replacing the outliers with its upper limit and lower limit This approach retains some information about the outliers while reducing their impact on the analysis.

```python
df['amount_in_usd_log'] = np.where(
    df['amount_in_usd_log'] > upper_limit,
    upper_limit,
    np.where(
        df['amount_in_usd_log'] < lower_limit,
        lower_limit,
        df['amount_in_usd_log']
    )
)
```

## Using Last 7 days sales as a feature into our data frame

```python
def difference_in_days(melt, lags, ffday, customer_id_col, create_date_col, net_amount_col):
    for i in range(ffday, lags+1):
        melt['Last-'+str(i)+'day_Sales'] = melt.groupby([customer_id_col])[net_amount_col].shift(i)
    
    melt = melt.reset_index(drop = True)
    
    for i in range(ffday, lags+1):
        melt['Last-'+str(i)+'day_Diff']  = melt.groupby([customer_id_col])['Last-'+str(i)+'day_Sales'].diff()
    melt = melt.fillna(0)
    return melt
new_df=difference_in_days(new_df,15,1,'unique_cust_id','ORDER_CREATION_DATE','amount_in_usd_log')
```

```python
from hyperopt import hp
space={'max_depth': hp.quniform("max_depth", 3, 18, 1),
        'gamma': hp.uniform ('gamma', 1,9),
        'reg_alpha' : hp.quniform('reg_alpha', 40,180,1),
        'reg_lambda' : hp.uniform('reg_lambda', 0,1),
        'colsample_bytree' : hp.uniform('colsample_bytree', 0.5,1),
        'min_child_weight' : hp.quniform('min_child_weight', 0, 10, 1),
        'n_estimators': 180,
        'seed': 0
    }
```

```python
import xgboost as xgb
from sklearn.metrics import mean_squared_error
from hyperopt import fmin, tpe, hp, STATUS_OK, Trials

def objective(space):
    clf = xgb.XGBRegressor(
        n_estimators=int(space['n_estimators']),
        max_depth=int(space['max_depth']),
        gamma=space['gamma'],
        reg_alpha=int(space['reg_alpha']),
        min_child_weight=int(space['min_child_weight']),
        colsample_bytree=space['colsample_bytree']
    )
    
    evaluation = [(X_train_sel, y_train), (X_test_sel, y_test)]
    
    clf.fit(X_train_sel, y_train,
            eval_set=evaluation, eval_metric="rmse",
            early_stopping_rounds=10, verbose=False)
    
    pred = clf.predict(X_test_sel)
    
    mse = mean_squared_error(y_test, pred)
    print("Mean Squared Error:", mse)
    
    return {'loss': mse, 'status': STATUS_OK}

# Define the search space for hyperparameters
space = {
    'n_estimators': hp.choice('n_estimators', [100, 200, 300]),
    'max_depth': hp.choice('max_depth', [3, 5, 7]),
    'gamma': hp.uniform('gamma', 0, 1),
    'reg_alpha': hp.choice('reg_alpha', [0, 1, 10]),
    'min_child_weight': hp.choice('min_child_weight', [1, 5, 10]),
    'colsample_bytree': hp.uniform('colsample_bytree', 0.5, 1)
}

trials = Trials()

best_hyperparams = fmin(fn=objective,
                        space=space,
                        algo=tpe.suggest,
                        max_evals=10,
                        trials=trials)

print("Best Hyperparameters:", best_hyperparams)
```
